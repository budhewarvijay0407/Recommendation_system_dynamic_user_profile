```python
#####Adding gaurdian links in the dataframe######
import requests
from bs4 import BeautifulSoup
month_lisr=['june','july','aug','sep']
date_list=np.arange(1,30)
final_news_portfolio=[]
#url='https://www.theguardian.com/sport/2022/sep/01/all'
for i in date_list:
    url_sports = 'https://www.theguardian.com/sport/2022/'+str(random.choice(month_lisr))+'/'+str(i).zfill(2)+'/'+'all'
    url_lifestyle = 'https://www.theguardian.com/lifeandstyle/2022/'+str(random.choice(month_lisr))+'/'+str(i).zfill(2)+'/'+'all'
    url_env='https://www.theguardian.com/environment/2022/'+str(random.choice(month_lisr))+'/'+str(i).zfill(2)+'/'+'all'
    url_business='https://www.theguardian.com/business/2022/'+str(random.choice(month_lisr))+'/'+str(i).zfill(2)+'/'+'all'
    url_fashion='https://www.theguardian.com/fashion/2022/'+str(random.choice(month_lisr))+'/'+str(i).zfill(2)+'/'+'all'
    final_news_portfolio.append(url_sports)
    final_news_portfolio.append(url_lifestyle)
    final_news_portfolio.append(url_env)
    final_news_portfolio.append(url_business)
    final_news_portfolio.append(url_fashion)

def fetch_unique_urls(url):
    reqs = requests.get(url)
    soup = BeautifulSoup(reqs.text, 'html.parser')

    unique_urls = []
    for link in soup.find_all('a'):
        url_test=link.get('href')
        if type(url_test)==str:
            if url[:-3] in url_test:
                unique_urls.append(url_test)
    unique_urls=list(set(unique_urls))
    return(unique_urls)

total_url=[]
for ur in final_news_portfolio:
    total_url.append(fetch_unique_urls(ur))
flat_list_url = [item for sublist in total_url for item in sublist]
```

```python
category=['sports','lifestyle','env','business']
whole_sports_url=[url for url in flat_list_url if 'sport' in url]
whole_lifestyle_url=[url for url in flat_list_url if 'lifeandstyle' in url]
whole_env_url=[url  for url in flat_list_url if 'environment' in url]
whole_business_url=[url for url in flat_list_url if 'business' in url]
#whole_url_fashion=[url for url in url_fashion if 'fashion' in url]
ip_list=final_df['IP'].unique()
category=[whole_sports_url,whole_lifestyle_url,whole_env_url,whole_business_url]
final_data=[]
for x in ip_list:
    df=final_df.loc[final_df['IP']==x]
    category_for_this_ip=random.choice(category)
    url_list=[url for url in random.choices(category_for_this_ip,k=len(df))]
    df['URL']=url_list
    final_data.append(df)
df_final=pd.concat(final_data)
```

```python
df_final.to_csv('sample_guardian_data.csv')
```
